```python
import pandas as pd
import numpy as np
```

```python
tpfilterPath = [
    r'/home/rbarreror/projects/TurboPutative-web/src/public/jobs/123456_pos/tsv/4_TPFilter.tsv',
    r'/home/rbarreror/projects/TurboPutative-web/src/public/jobs/123456_neg/tsv/4_TPFilter.tsv'
]

tpf = [pd.read_csv(i, sep='\t', low_memory=False, index_col=0) for i in tpfilterPath]
```

```python
m2iPath = r"/home/rbarreror/projects/TurboPutative-web/src/TurboOmicsIntegrator/jobs/123456/EDA/xPreProcessing/m2i.json"
m2i = pd.read_json(m2iPath)
m2i = m2i.set_index(m2i.columns[0])
```

```python
for i in range(len(tpf)):
    tpf[i] = tpf[i].iloc[:, m2i.shape[1]:]
    tpf[i] = tpf[i][[col for col in tpf[i].columns if '_quant_' not in col]]
    tpf[i].columns = [f'TP_{col}' if 'TP_' not in col else col for col in tpf[i]]
    tpf[i]['_id'] = tpf[i].index
    tpf[i] = tpf[i].fillna('')
    tpf[i] = tpf[i].groupby('_id').agg(lambda x: ' & '.join([str(n) for n in x.tolist()]))
```

```python
COLUMNS = ['TP_Name', 'TP_Adduct', 'TP_mz Error (ppm)', 'TP_Class', 'TP_TPMetrics']
```

```python
m2itp = m2i.join(tpf[0]).fillna('')
```

```python
m2itp[m2i.columns.tolist()+COLUMNS].to_json(m2iPath, orient='records')
```

```python
m2iTP = r"/home/rbarreror/projects/TurboPutative-web/src/TurboOmicsIntegrator/jobs/123456/EDA/xPreProcessing/m2iTP.json"
m2itp.to_json(m2iTP, orient='records')
```
